```python
from collections import defaultdict

f = open("dataset_865453_5.txt","r")

s1, s2, s3 = f.readline().strip('\n'), f.readline().strip('\n'), f.readline().strip('\n')
f.close()

def MultipleAlignment(s1, s2, s3):
    l, m, n = len(s1) + 1, len(s2) + 1, len(s3) + 1
    DAG = defaultdict(lambda: ((-1,-1,-1),0))
    
    DAG[(0,0,0)] = ((-1,-1,-1),0)
    for i in range(1,l):
        DAG[(i,0,0)] = ((i-1,0,0),0)
    for j in range(1,m):
        DAG[(0,j,0)] = ((0,j-1,0),0)
    for k in range(1,n):
        DAG[(0,0,k)] = ((0,0,k-1),0)
    
    for i in range(1,l):
        for j in range(1,m):
            for k in range(1,n):
                    
                match = 1 if s1[i-1] == s2[j-1] and s1[i-1] == s3[k-1] else 0
                max_incoming = max(DAG[(i-1,j,k)][1], DAG[(i,j-1,k)][1], DAG[(i,j,k-1)][1], \
                    DAG[(i-1,j-1,k)][1], DAG[(i-1,j,k-1)][1], DAG[(i,j-1,k-1)][1], DAG[(i-1,j-1,k-1)][1] + match)
                
                incoming_edge = (-1, -1, -1)
                if max_incoming == DAG[(i-1,j-1,k-1)][1] + match:
                    incoming_edge = (i-1,j-1,k-1)
                elif max_incoming == DAG[(i-1,j,k)][1]:
                    incoming_edge = (i-1,j,k)
                elif max_incoming == DAG[(i,j-1,k)][1]:
                    incoming_edge = (i,j-1,k)
                elif max_incoming == DAG[(i,j,k-1)][1]:
                    incoming_edge = (i,j,k-1)
                elif max_incoming == DAG[(i-1,j-1,k)][1]:
                    incoming_edge = (i-1,j-1,k)
                elif max_incoming == DAG[(i-1,j,k-1)][1]:
                    incoming_edge = (i-1,j,k-1)
                else:
                    incoming_edge = (i,j-1,k-1)
                    
                DAG[(i,j,k)] = (incoming_edge, max_incoming)
    
    
    curr_node = (l-1, m-1, n-1)
    last_node = (-1,-1,-1)
    end_length = DAG[curr_node][1]
    a1, a2, a3 = "", "", ""
    while curr_node != (-1,-1,-1):
        prev_node = DAG[curr_node][0]
        if prev_node == (-1,-1,-1):
            break
            
        if prev_node[0] < curr_node[0]:
            a1 = s1[curr_node[0] - 1] + a1
        else: 
            a1 = '-' + a1
        if prev_node[1] < curr_node[1]:
            a2 = s2[curr_node[1] - 1] + a2
        else: 
            a2 = '-' + a2
        if prev_node[2] < curr_node[2]:
            a3 = s3[curr_node[2] - 1] + a3
        else: 
            a3 = '-' + a3
        
        curr_node = prev_node
    
    while curr_node[0] > 0:
        a1 = s1[curr_node[0] - 1] + a1
        curr_node = (curr_node[0] - 1, curr_node[1], curr_node[2])
    while curr_node[1] > 0:
        a2 = s2[curr_node[1] - 1] + a2
        curr_node = (curr_node[0], curr_node[1] - 1, curr_node[2])
    while curr_node[2] > 0:
        a3 = s3[curr_node[2] - 1] + a3
        curr_node = (curr_node[0], curr_node[1], curr_node[2] - 1)
    
    aligned_length = max(len(a1), len(a2), len(a3))
    while len(a1) < aligned_length:
        a1 = '-' + a1
    while len(a2) < aligned_length:
        a2 = '-' + a2
    while len(a3) < aligned_length:
        a3 = '-' + a3
    
    return end_length, a1, a2, a3

for item in MultipleAlignment(s1, s2, s3):
    print(item)
```
